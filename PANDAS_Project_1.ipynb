```python
import pandas as pd
xl = pd.ExcelFile("/content/TableauSalesData.xlsx")
SalesData = xl.parse("Orders")
```

```python
display(SalesData.head(10))
```

```python
RegionData = SalesData[["Region", "Sales", "Profit"]].groupby(by="Region").sum().sort_values(by="Sales", ascending = False)
display(RegionData)
display('Table 1: Least profitable Regions')
```

```python
SalesTrends = SalesData[["Category", "Region", "Sales", "Profit","Segment"]]
SouthTrends = SalesTrends.loc[(SalesTrends["Region"]=="South")].groupby(by= "Category").sum().sort_values(by="Sales",ascending = True)
display(SouthTrends)
```

```python
SouthCategory= SalesData[["Category","Sub-Category","Region", "Sales","Profit"]]
SouthFurniture = SouthCategory.loc[(SouthCategory["Category"]=="Furniture")& (SouthCategory["Region"]=="South")]
TableProf = SouthFurniture.groupby(["Category", "Sub-Category"]).sum().sort_values(by="Profit", ascending = False)
display(TableProf)
```

Above we can see that the South Region has the least amount of sales, now we need to check if they are ordering, what kind of discounts we are giving, and the shipping rate.We need to break it down by state. First we will breakdown the south region by Sub-Category and see what are the strongest and weakest items.

```python
SegmentSales = SalesData[["Segment","Region" ,"State","Sub-Category","Sales", "Profit"]]
SouthSegment = SegmentSales.loc[(SegmentSales["Region"]=="South") & (SegmentSales["Sub-Category"]=="Tables")]
SegmentProf = SouthSegment.groupby(by="State").sum().sort_values(by="Profit", ascending = False)
display(SegmentProf)
print(" Table 2: Southern Region States by")
print("              Sales and Profit     ")
```

```python
SubCatProfSales = SalesData[["Region", "Category","Sub-Category", "Sales", "Profit", "Ship Mode"]]
SouthProfSales = SubCatProfSales.loc[(SubCatProfSales["Region"]=="South") & (SubCatProfSales["Sub-Category"]=="Tables")]
TotalSouthProfSales = SouthProfSales.groupby(["Ship Mode", "Sub-Category"]).sum().sort_values(by="Ship Mode", ascending = False)
display(TotalSouthProfSales)
```

```python
SegmentSales = SalesData[["Segment","Region" ,"State","Sub-Category", "Ship Mode", "Profit", "Sales"]]
SouthSegment = SegmentSales.loc[(SegmentSales["State"]=="North Carolina") & (SegmentSales["Sub-Category"]=="Tables")]
SegmentProf = SouthSegment.groupby(["State","Ship Mode"]).sum().sort_values(by="Profit", ascending = True)
display(SegmentProf)
```

```python
SegmentSales = SalesData[["Segment","Region" ,"State","Sub-Category", "Ship Mode", "Profit", "Sales", "Discount"]]
SouthSegment = SegmentSales.loc[(SegmentSales["State"]=="North Carolina") & (SegmentSales["Sub-Category"]=="Tables")]
SegmentProf = SouthSegment.groupby(["State","Ship Mode"]).sum().sort_values(by="Discount")
display(SegmentProf)
```

```python
NorCarTables = SalesData[["State", "Profit", "Discount"]]
NorCarState = NorCarTables.loc[(NorCarTables["State"]=="North Carolina")]
TotalAveDiscount = NorCarState.groupby("State").agg({"Profit":'sum' , "Discount":'mean'})
display (TotalAveDiscount)
```

```python
SegmentSales = SalesData.loc[SalesData["Sub-Category"]=="Tables"]
SegmentSalesYear = SegmentSales.copy()
SegmentSalesYear["Year"] = SegmentSalesYear["Order Date"].dt.year
Years = SegmentSalesYear.Year.unique()
SegmentTableYear = SegmentSalesYear[["State", "Year", "Profit", "Sales", "Discount"]]
StateTableYear = SegmentTableYear.loc[SegmentTableYear["State"]=="North Carolina"]
for year in Years:
  OneYear = StateTableYear.loc[StateTableYear["Year"]==year]
  NoYear = OneYear[[ "Discount", "Profit", "Sales"]]
  YearlyDiscount = StateTableYear.groupby(by="State").mean().sort_values(by="Profit")
  print("\nTable Discounts for the year " + str(year) + " in North Carolina")
  display(YearlyDiscount)
  display("*"*40)
```

```python
SegmentSales = SalesData.loc[SalesData["Sub-Category"]=="Tables"]
SegmentSalesMonth = SegmentSales.copy()
SegmentSalesMonth["Month"] = SegmentSalesMonth["Order Date"].dt.month
Months = SegmentSalesMonth.Month.unique()
SegmentTableMonth = SegmentSalesMonth[["Sub-Category", "Month", "Sales", "Profit", "Discount"]]
for month in Months:
  OneMonth= SegmentTableMonth.loc[SegmentTableMonth["Sub-Category"]=="Tables"]
  NoMonth = SegmentTableMonth[[ "Sub-Category", "Sales","Month","Profit", "Discount"]]
  MonthlyDiscount = NoMonth.groupby(["Sub-Category","Month"]).mean().sort_values(by="Month",ascending = True)
  display(MonthlyDiscount)
  print("Table 3: Average Discount provided for each month")
```
